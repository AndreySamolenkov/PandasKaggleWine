### Load the dataset into a Pandas dataframe and display the first 5 rows.

```python
import pandas as pd

df = pd.read_csv('winemag-data_first150k.csv')
df.head(5)
```

### Display the number of rows and columns in the dataset.

```python
df.shape
```

### Check if there are any missing values in the dataset.

```python
df.isnull().sum()
```

### Display the number of unique values in the 'country' column.

```python
df['country'].nunique()
```

### Display the mean, median, and mode of the 'points' column.

```python
points_mean = df['points'].mean()
points_mean
```

```python
points_median = df['points'].median()
points_median
```

```python
points_mode = df['points'].mode()[0]
points_mode
```

### Display the top 5 most frequent values in the 'variety' column.

```python
most_freq = df['variety'].value_counts()
most_freq.head(5)
```

### Group the data by 'country' and calculate the mean 'points' for each country.

```python
country_points_mean = df.groupby(['country'])['points'].mean()
country_points_mean
```

### Sort the data by 'price' in descending order.

```python
sort_price = df.sort_values('price', ascending=False)
sort_price.head(10)
```

### Display the 10 wines with the highest 'points' and 'price'.

```python
display_wines = df[['points', 'price']].sort_values(['points', 'price'], ascending=[False, False]).head(10)
display_wines
```

### Create a new column called 'price_category' that categorizes wines based on their price as 'cheap', 'moderate', or 'expensive'.

```python
def price_category(price):
    if price < 20:
        return 'Cheap'
    elif price < 50:
        return 'Moderate'
    else:
        return 'Expensive'
    
df['price_category'] = df['price'].apply(price_category)

df.head()
```

### Display the number of wines in each price category.

```python
price_cat_nums = df['price_category'].value_counts()
price_cat_nums
```

### Display the mean 'points' for each price category.

```python
df.groupby(['price_category'])['points'].mean()
```

### Create a new dataframe that only includes wines from the 'US' and 'France'.

```python
us_france_wines = df.loc[df['country'].isin(['US', 'France'])]
us_france_wines
```

### Display the number of wines from each country in the new dataframe.

```python
us_france_wines['country'].value_counts()
```

### Calculate the mean 'points' for each country in the new dataframe.

```python
us_france_wines.groupby(['country'])['points'].mean()
```

### Create a scatter plot showing the relationship between 'price' and 'points'.

```python
import matplotlib.pyplot as plt

plt.scatter(df['price'], df['points'])
plt.xlabel('Price')
plt.ylabel('Points')
plt.title('Price vs Points')
plt.show()
```

### Create a bar plot showing the number of wines from each country.

```python
country_counts = df['country'].value_counts()
plt.bar(country_counts.index, country_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of Wines')
plt.title('Number of Wines by Country')
plt.show()
```

### Create a histogram showing the distribution of 'points'.

```python
plt.hist(df['points'], bins=20)
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.title('Distribution of Points')
plt.show()
```

### Create a box plot showing the distribution of 'points' for each country.

```python
df.boxplot(column='points', by='country', rot=90)
plt.xlabel('Country')
plt.ylabel('Points')
plt.title('Distribution of Points by Country')
plt.show()
```

### Create a pivot table that shows the mean 'points' for each country and 'variety' combination.

```python
pivot_table = df.pivot_table(values='points', index='country', columns='variety', aggfunc='mean')
pivot_table
```
